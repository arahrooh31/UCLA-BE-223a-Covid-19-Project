## Cleaning City Names in hoodCOVID19DF

```python
#Removes "City of", "Unincorporated -", and "Los Angeles -" prefixes in hoodCOVID19DF so the place_IDs can match 
#the NEIGHBORHOODS in nbhdAges
city_of = "City of "
unincorporated = "Unincorporated "
LosAngeles = "Los Angeles - "
for i, row in hoodCOVID19DF.iterrows():
    if city_of in row[0]:
        new_name = row[0].replace(city_of, '')
        hoodCOVID19DF.at[i, 'place_ID'] = new_name
    elif unincorporated in row[0]:
        new_name = row[0].replace(unincorporated, '')
        hoodCOVID19DF.at[i, 'place_ID'] = new_name
    elif LosAngeles in row[0]:
        new_name = row[0].replace(LosAngeles, '')
        hoodCOVID19DF.at[i, 'place_ID'] = new_name

hoodCOVID19DF
```

```python

##########################################################################################################################
#Declaring variables
totalR = 0
R = []
places = []
pop18OrLess = []
pop19To49 = []
pop50To64 = []
pop65AndUp = []
Totalpopulation = []
casePrcnt_18OrLess = []
casePrcnt_19To49 = []
casePrcnt_50To64 = []
casePrcnt_50To64 = []
casePrcnt_65AndUp = []
sumOfPlaceCase = []

##########################################################################################################################
#Defining some variables for the columns of combinedHoodDF
for i in range(1,len(nbhdAgesDF)+1):
    places.append(nbhdAgesDF['NEIGHBORHOOD'][i])
    Totalpopulation.append(nbhdAgesDF['POPULATION_TOTAL'][i])
    pop18OrLess.append(nbhdAgesDF['AGE_18_OR_LESS'][i])
    pop19To49.append(nbhdAgesDF['AGE_19_TO_49'][i])
    pop50To64.append(nbhdAgesDF['AGE_50_TO_64'][i])
    pop65AndUp.append(nbhdAgesDF['AGE_65_AND_UP'][i])
    totalR += 1

##########################################################################################################################
#Interpolating percent of infections by age categories based on LA County infected percents, LA County demographics, 
#and the Places' demographics
for i in range(0,totalR):
    casePrcnt_18OrLess.append(pop18OrLess[i] * combinedLACOVID19DF['%OfTotalCases'][0] / LA_CombinedAgePercents['AdjustedPopulationPercent'][0])
    casePrcnt_19To49.append(pop19To49[i] * combinedLACOVID19DF['%OfTotalCases'][1] / LA_CombinedAgePercents['AdjustedPopulationPercent'][1])
    casePrcnt_50To64.append(pop50To64[i] * combinedLACOVID19DF['%OfTotalCases'][2] / LA_CombinedAgePercents['AdjustedPopulationPercent'][2])
    casePrcnt_65AndUp.append(pop65AndUp[i] * combinedLACOVID19DF['%OfTotalCases'][3] / LA_CombinedAgePercents['AdjustedPopulationPercent'][3])
    sumOfPlaceCase.append(casePrcnt_18OrLess[i] + casePrcnt_19To49[i] + casePrcnt_50To64[i] + casePrcnt_65AndUp[i]) #Should be 1

##########################################################################################################################
#Printing some results
print(totalR, " places have demographic and COVID19 data")
print(len(hoodCOVID19DF)-totalR, " places only have COVID19 data and not demographic data")
combinedHoodDF = pd.DataFrame({'Place':places, 'TotalPopulation':Totalpopulation, 
                               'pop18OrLess':pop18OrLess, 'pop19To49':pop19To49, 'pop50To64':pop50To64, 'pop65AndUp':pop65AndUp, 
                               'casePrcnt_18OrLess':casePrcnt_18OrLess, 'casePrcnt_19To49':casePrcnt_19To49, 'casePrcnt_50To64':casePrcnt_50To64, 'casePrcnt_65AndUp':casePrcnt_65AndUp, 'sumOfPlaceCasePercents':sumOfPlaceCase})

##########################################################################################################################
#Interpolating the age demographics of the infected cases: combinedHoodDF
for r in range(0,len(combinedHoodDF)):
    combinedHoodDF['casePrcnt_18OrLess'][r] = combinedHoodDF['casePrcnt_18OrLess'][r]/combinedHoodDF['sumOfPlaceCasePercents'][r]
    combinedHoodDF['casePrcnt_19To49'][r] = combinedHoodDF['casePrcnt_19To49'][r]/combinedHoodDF['sumOfPlaceCasePercents'][r]
    combinedHoodDF['casePrcnt_50To64'][r] = combinedHoodDF['casePrcnt_50To64'][r]/combinedHoodDF['sumOfPlaceCasePercents'][r]
    combinedHoodDF['casePrcnt_65AndUp'][r] = combinedHoodDF['casePrcnt_65AndUp'][r]/combinedHoodDF['sumOfPlaceCasePercents'][r]
    combinedHoodDF['sumOfPlaceCasePercents'][r] = combinedHoodDF['casePrcnt_18OrLess'][r] + combinedHoodDF['casePrcnt_19To49'][r] + combinedHoodDF['casePrcnt_50To64'][r] + combinedHoodDF['casePrcnt_65AndUp'][r] #Should be 1
combinedHoodDF
```

```python
#Prepare the dataframe for interpolating the percent of *age category* infected per each place.
#Add the cases for each city to the combinedHoodDF dataframe. Take data from the hoodCOVID19DF dataframe. 
#Find the combinedHoodDF's place's corresponding row in the COVID19DF. If the array of the indices is not empty, 
#Add the Cases entry to the corresponding row in the combinedHoodDF.

#Note: Takes first matching element, second one is probably from what was "Unincorporated [placename]" before cleaning

#Makes a 'Cases' column of zeroes for the combinedHoodDF 
zeroes = []
for i in range(0,len(combinedHoodDF)):
    zeroes.append(0)
combinedHoodDF['Cases'] = zeroes

#Adds the 'Cases' for each place when it is known. Adds 'NA' if Cases are unknown
for i in range(0,len(combinedHoodDF)):
    #Finds the corresponding row in the COVID19 data for a place
    correspondingRow = (hoodCOVID19DF[hoodCOVID19DF['place_ID']==combinedHoodDF['Place'][i]].index.values)
    #If there is COVID data for a place, add the Cases to the combined DF
    if correspondingRow.size > 0: 
        combinedHoodDF['Cases'][i] = hoodCOVID19DF['cumulative_cases_today'][correspondingRow[0]]
    #If there is no COVID data for a place, make a really big number for the Cases in the combined DF
    else: 
        combinedHoodDF['Cases'][i] = np.nan
        
combinedHoodDF
```

## Rina's comments

- %ofxxtoxxPopInfected is not correct.
- For example, for of18OrLessInfected, all 3 variables used in the calculation, are purely from community level.
- combinedHoodDF['pop18OrLess'] is just the percent of people who are under 18, combinedHoodDF['Cases'] divided by combinedHoodDF['TotalPopulation'] simply gets you the percentage of people infected in that community.
- So multiplying the percentage of people under 18 with percentage of people infected in that community won't let us know the percentage of people who are both "1)under 18 years old and 2) infected", at the same time.

```python
#Declaring variables
of18OrLessInfected = []
of19to49Infected = []
of50to64Infected = []
of65andUpInfected = []


#Interpolate/estimate the percent of an Age Category that is infected per each Place (estimated percent of cases which are in age category 'i' * # cases / population)
for i in range(0,len(combinedHoodDF)):
    #If the number of cases are known for a place, estimate the percent of an Age Category that is infected
    if combinedHoodDF['Cases'] is not np.nan:
        of18OrLessInfected.append(combinedHoodDF['pop18OrLess'][i]*(combinedHoodDF['Cases'][i]/combinedHoodDF['TotalPopulation'][i]))  
        of19to49Infected.append(combinedHoodDF['pop19To49'][i]*combinedHoodDF['Cases'][i]/combinedHoodDF['TotalPopulation'][i])
        of50to64Infected.append(combinedHoodDF['pop50To64'][i]*combinedHoodDF['Cases'][i]/combinedHoodDF['TotalPopulation'][i])
        of65andUpInfected.append(combinedHoodDF['pop65AndUp'][i]*combinedHoodDF['Cases'][i]/combinedHoodDF['TotalPopulation'][i])
    #If the number of cases is not known, assign 'not a number'
    else:
        of18OrLessInfected.append(np.nan)  
        of19to49Infected.append(np.nan)
        of50to64Infected.append(np.nan)
        of65andUpInfected.append(np.nan)

#Just assigns the arrays to the dataframe
combinedHoodDF['%Of18orLessPopInfected'] = of18OrLessInfected  
combinedHoodDF['%Of19to49PopInfected'] = of19to49Infected
combinedHoodDF['%Of50to64PopInfected'] = of50to64Infected
combinedHoodDF['%Of65andUpPopInfected'] = of65andUpInfected

combinedHoodDF
```
